```python
def makeSpeeds(num_edges, num_intervals, endTime,filename):
    speeds = [30,40,50,60,70,80]
    s= str(num_intervals) +' '+ str(endTime) + '\n'
    for i in range(num_edges):
        dist = random.randint(1,10)
        s+=str(dist)
        for n in range(num_intervals):
            speed = speeds[random.randint(0,5)]
            s += ' ' + str(speed)
        s += '\n'
    fo = open(filename,'w')
    fo.write(s)
    fo.close()
```

```python
def make_clusters(m,n,filename):
    fo = open(filename,'w')
    for i in range(n*m*4):
        fo.write(str(i/(n*m))+"\n")
#     for i in range(101,109):
#         if(i==101):
#             fo.write
    s = '0\n1\n0\n1\n2\n3\n2\n3'
    fo.write(s)
    fo.close()
```
